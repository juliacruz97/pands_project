```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Carregar o dataset Iris e criar um DataFrame
data = load_iris()
iris = pd.DataFrame(data.data, columns=data.feature_names)

# Criar uma funÃ§Ã£o para mapear os valores de x e y para diferentes cores
def color_points(row):
    if row['sepal length (cm)'] > 5 and row['sepal width (cm)'] > 3:
        return 'red'
    elif row['sepal length (cm)'] <= 5 and row['sepal width (cm)'] > 3:
        return 'blue'
    elif row['sepal length (cm)'] > 5 and row['sepal width (cm)'] <= 3:
        return 'green'
    else:
        return 'purple'

# Aplicar a funÃ§Ã£o ao DataFrame para criar uma nova coluna de cores
iris['color'] = iris.apply(color_points, axis=1)

# Criar o grÃ¡fico de dispersÃ£o com cores diferentes para cada ponto
scatter = iris.plot.scatter(x='sepal length (cm)', y='sepal width (cm)', c=iris['color'], edgecolor='black')

# Add title and labels to axes
scatter.set_title('Scatter Plot ', fontsize=14)
scatter.set_xlabel('Sepal Length (cm)', fontsize=12)
scatter.set_ylabel('Sepal Width (cm)', fontsize=12)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the Iris dataset and create a DataFrame
data = load_iris()
iris = pd.DataFrame(data.data, columns=data.feature_names)

# Calculate the correlation matrix
correlation_matrix = iris.corr()

# Create the correlation graph (heatmap)
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title('Correlation Matrix ', fontsize=14)

plt.show()
```
