```python
import pandas as pd
import matplotlib.pyplot as plt

# Exemplo de DataFrame com dados categóricos
data = {'species': ['setosa','versicolor', 'virginoca',]}
iris = pd.DataFrame(data)
```

```python
# Converter categorias em números
iris['species_cod'] = pd.Categorical(iris['species']).codes
```

```python

# Criar o histograma usando os códigos numéricos
iris['species_cod'].plot(kind='hist', bins=3, rwidth=0.8)
plt.xlabel('Code of Species')
plt.ylabel('Frequêncy')
plt.title('Histogram of Species')
plt.show()
```
